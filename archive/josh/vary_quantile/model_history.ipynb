```python
import pickle as pickle
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from keras.models import load_model
```

# model accuracy during training

```python
def plot_history(filenames):
    history = []
    for filename in filenames:
        with open(filename, 'rb') as pickle_file:
            history.append(pickle.load(pickle_file))
            
    for idx, hist in enumerate(history):
        # summarize history for accuracy
        #plt.plot(hist['acc'], label=f'{filename[:5]}')
        plt.plot(hist['val_acc'], label=f'{filenames[idx][:5]}')
        plt.title('model accuracy')
        plt.ylabel('accuracy')
        plt.xlabel('epoch')
        plt.legend(loc='upper left')
    plt.show()
    for idx, hist in enumerate(history):
        # summarize history for loss
        #plt.plot(hist['loss'])
        plt.plot(hist['val_loss'], label=f'{filenames[idx][:5]}')
        plt.title('model loss')
        plt.ylabel('loss')
        plt.xlabel('epoch')
        plt.legend(loc='upper right')
    plt.show()
```

```python
plot_history(['10_90_history.pkl', '20_80_history.pkl', 
              '30_70_history.pkl', '40_60_history.pkl',
              '50_50_history.pkl'])
```
